## 3分析步骤

### 3.1csv文件的读取
对csv文件的读取

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif']=['simHei']
shixi=pd.read_csv(r"C:\\Users\\tgq\\Desktop\\实习生\\shixiseng\\mlalgorithm.csv", delimiter=',', encoding='utf8', error_bad_lines=False,names=['auth_capital','city','com_class','com_fullname','com_id','com_intro','com_links','com_location','com_logo','com_name','com_website','com_welfare','day_per_week','detailed_intro','est_date','industry','job_academic','job_deadline','job_detail','job_links','job_title','num_employee','released_time','tag','time_span','update_time','wage'])
shixi=shixi.replace(' ',',')
shixi
```

### 3.2数据清洗
对提取的数据进行去重、规范化处理

```python
#提取有效的列
shixi1=shixi.loc[:,['auth_capital','city','com_fullname','com_location','com_name']]
shixi1.head()
```

```python
#去重
shixi = shixi.drop_duplicates().reset_index().drop('index',axis=1)
shixi
```

```python
#各个城市的招聘岗位数量
shixi=shixi.replace(',上海','上海')
citysum=shixi.groupby('city')['com_fullname'].count()
citysum=citysum.reset_index().sort_values(by='com_fullname',ascending=False)
citysum1=citysum.head()
citysum
```

```python
#各个城市的招聘岗位数量（图表）
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(1,1,1)
ax.bar(citysum['city'],citysum['com_fullname'])
```

```python
#各个公司的注册资本
shixi2=shixi.loc[:,['auth_capital','com_fullname']]
shixi2
```

```python
#各个公司的招聘要求
citytag=shixi.groupby(['city','tag'])['com_fullname'].count()
citytag=citytag.reset_index().sort_values(by='tag',ascending=False)
citytag1=citytag.head()
citytag
```

```python
#各个城市的岗位需求（图表）
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(1,1,1)
ax.bar(citytag['tag'],citytag['com_fullname'])
```
