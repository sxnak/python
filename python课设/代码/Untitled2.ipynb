## 3分析步骤

### 3.1csv文件的读取
对csv文件的读取

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif']=['simHei']
shixi=pd.read_csv(r"C:\\Users\\tgq\\Desktop\\实习生\\shixiseng_packet_capture\\com_detailed.csv", delimiter=',', encoding='utf8',error_bad_lines=False,names=['address','cname','com_type','com_url','description','industry','info','is_collect','logo','name','pranum','reg_capi','reg_name','reg_num','scale','start_time','tags','types','url','vote_url','wurl'])
shixi=shixi.replace(' ',',')
shixi
```

### 3.2数据清洗
对提取的数据进行去重、规范化处理

```python
#提取有效的列
shixi1=shixi.loc[:,['description','address','cname','com_type','reg_capi']]
shixi1.head()
```

```python
#去重
shixi = shixi.drop_duplicates().reset_index().drop('index',axis=1)
shixi
```

### 3.3 数据分析
对进过处理的有效数据进行分析

```python
#各个城市的招聘岗位数量
shixi=shixi.replace(',上海','上海')
citysum=shixi.groupby('address')['cname'].count()
citysum=citysum.reset_index().sort_values(by='cname',ascending=False)
citysum1=citysum.head()
citysum
```

```python
#各个城市的招聘岗位数量（图表）
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(1,1,1)
ax.bar(citysum['address'],citysum['cname'])
```

```python
#对公司类型进行统计
citytype=shixi.groupby('com_type')['cname'].count()
citytype=citytype.reset_index().sort_values(by='cname',ascending=False)
citytype1=citytype.head()
citytype
```

```python
#对公司类型进行统计（图表）
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(1,1,1)
ax.bar(citytype1['com_type'],citytype1['cname'])
```

```python
#各个城市公司的产业
cityind=shixi.groupby(['address','industry'])['cname'].count()
cityind=cityind.reset_index().sort_values(by='cname',ascending=False)
cityind1=cityind.head(18)
cityind
```

```python
#统计全国公司产业（图表)
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(1,1,1)
ax.bar(cityind1['industry'],cityind1['cname'])
```
