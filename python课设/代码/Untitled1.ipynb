## 3分析步骤

### 3.1csv文件的读取
对csv文件的读取

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
plt.rcParams['font.sans-serif']=['simHei']
shixi=pd.read_csv(r"C:\\Users\\tgq\\Desktop\\实习生\\shixiseng_packet_capture\\job_list.csv", delimiter=',', encoding='utf8',error_bad_lines=False,names=['ad_type','city','cname','day','deliver','ftype','invite','maxsal','minsal','name','refresh','type','url','uuid'])
shixi=shixi.replace(' ',',')
shixi
```

### 3.2数据清洗
对提取的数据进行去重、规范化处理

```python
#提取有效的列
shixi1=shixi.loc[:,['city','day','name','maxsal','minsal']]
shixi1.head()
```

```python
#去重
shixi = shixi.drop_duplicates().reset_index().drop('index',axis=1)
shixi
```

```python
#各个城市的招聘岗位数量
shixi=shixi.replace(',上海','上海')
citysum=shixi.groupby('city')['cname'].count()
citysum=citysum.reset_index().sort_values(by='cname',ascending=False)
citysum1=citysum.head()
citysum
```

```python
#各个城市的招聘岗位数量（图表）
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(1,1,1)
ax.bar(citysum['city'],citysum['cname'])
```

```python
#统计各个城市的工作天数
cityday=shixi.groupby(['city','day'])['cname'].count()
cityday=cityday.reset_index().sort_values(by='cname',ascending=False)
cityday
```

```python
#统计全国招聘岗位工作天数的多少（图表)
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(3,3,3)
ax.bar(cityday['day'],cityday['cname'])
```

```python
#各个岗位的招聘要求
shixi1=shixi.loc[:,['name','cname']]
shixi1
```

```python
#各个城市岗位的工资待遇
citysalary=shixi.groupby(['city','maxsal','minsal'])['cname'].sum()
citysalary=citysalary.reset_index().sort_values(by='maxsal',ascending=False)
citysalary
```

```python
#计算各个城市的最高平均工资和最低平均工资
citysalary1=shixi.groupby(['city'])['maxsal','minsal'].mean()
citysalary1=citysalary1.reset_index().sort_values(by='maxsal',ascending=False)
citysalary1
```

```python
#各个城市岗位的平均最高工资（图表）
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(1,1,1)
ax.bar(citysalary1['city'],citysalary1['maxsal'])
```

```python
#各个城市岗位的平均最低工资（图表）
fig=plt.figure(figsize=(15,15))
ax=fig.add_subplot(1,1,1)
ax.bar(citysalary1['city'],citysalary1['minsal'])
```
